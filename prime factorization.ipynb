```python
from IPython.display import display, Math
import sympy as sym
import numpy as np
import math
```

```python
number = 48

factor_dict = sym.factorint(number) # computes prime factors. generates dictionary with prime factors

print(factor_dict.keys()) # displays keys, values
print(list(factor_dict)) # axtracts keys # converts to list
print(factor_dict)
```

```python
# exercise.indepentent

# number = int(input('what number?')) # input for testing
number = range(2,500)
prime = 'is a prime number'
comp = 'is a compisite number with prime factors'

for i in number:
    factor_dict = sym.factorint(i)

    listnum = list(factor_dict)

    if factor_dict.get(listnum[0]) < 2 and len(factor_dict) is 1:
        print(i,prime)
    else:
        print(i,comp,listnum)
```

```python
# exercise.instructed
nums = range(2,51)

for numi in nums:
    di = sym.factorint(numi)
    ks = list(di.keys())
    if len(di)==1 and di[ks[0]]==1:
        print('%g is a prime number' %(numi))
    else:
        print('%g is a composite number with prime factors: %s' %(numi,list(di.keys())))
```
